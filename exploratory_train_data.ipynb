```python
import os
import gc
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = "SimHei"
import seaborn as sns
```

```python
user_info = pd.read_csv(os.getcwd() + '/data_processing/train_user_info.csv')
```

```python
user_info.info()
```

```python
user_info.head()
```

```python
user_info['gender_cn'] = user_info['gender'].replace({0:'女性', 1:'男性', 2:'未知'})
user_info['gender_cn'].fillna('空值', inplace=True)

user_info['age_range_cn'] = user_info['age_range'].replace({0:'未知', 1: '<18', 2: '[18,24]', 3: '[25,29]', 4: '[30,34]', 5: '[35,39]', 6: '[40,49]', 7 :'[50,59]', 8:'>=60'})
user_info['age_range_cn'].fillna('空值', inplace=True)

user_info['label_cn'] = user_info['label'].replace({0:'非重复买家', 1:'重复买家'})
```

```python
pd.crosstab(user_info.gender_cn, user_info.age_range_cn, margins=True)
```

```python
pd.crosstab(user_info.label_cn, user_info.gender_cn, margins=True)
```

```python
pd.crosstab(user_info.label_cn, user_info.age_range_cn, margins=True)
```
